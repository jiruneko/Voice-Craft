```python
import os
import sys

sys.path.append("./audiocraft")

os.environ['USER'] = 'temp'
os.environ['HOME'] = '/content/VoiceCraft/temp'

import numpy as np
import random
import torch
import torchaudio

from huggingface_hub import hf_hub_download

from data.tokenizer import (
    AudioTokenizer,
    TextTokenizer,
)
from inference_tts_scale import inference_one_sample

from models import voicecraft

from yt_dlp import YoutubeDL
from moviepy.editor import VideoFileClip

from IPython.display import Audio

SEED = 12
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print('using device is', device)
```

```python
def seed_everything(seed):
  os.environ['PYTHONHASHSEED'] = str(seed)
  random.seed(seed)
  np.random.seed(seed)
  torch.manual_seed(seed)
  torch.cuda.manual_seed(seed)
  torch.backends.cudnn.benchmark = False
  torch.backends.cudnn.deterministic = True

seed_everything(SEED)
```

```python
# download url
video_url = 'https://www.youtube.com/watch?v=GE-7bswoCzg'

# clip section
start_sec =  0
end_sec1 =  10
end_sec2 = 4.2

(start_pt, end_pt1) = (start_sec, end_sec1)
(start_pt, end_pt2) = (start_sec, end_sec2)
```

```python
%cd /content/VoiceCraft
```

```python
!mkdir -p inputs/video
!mkdir -p inputs/audio

download_resolution = 360
full_video_path = 'inputs/video/full_video.mp4'
clip_video_path1 = 'inputs/video/clip_video1.mp4'
clip_video_path2 = 'inputs/video/clip_video2.mp4'


# download video
ydl_opts = {'format': f'best[height<={download_resolution}]', 'overwrites': True, 'outtmpl': full_video_path}
with YoutubeDL(ydl_opts) as ydl:
  ydl.download([video_url])

# clip video
with VideoFileClip(full_video_path) as video:
  subclip = video.subclip(start_pt, end_pt1)
  subclip.write_videofile(clip_video_path1)

with VideoFileClip(full_video_path) as video:
  subclip = video.subclip(start_pt, end_pt2)
  subclip.write_videofile(clip_video_path2)
```

```python
# load video using MoviePy
videoclip = VideoFileClip(clip_video_path1)

original_audio_path = 'inputs/audio/original_audio.wav'

# codecがトレーニングされているsampling rateに合わせてfpsを設定
codec_audio_sr = 16000

# extract audio
audioclip = videoclip.audio
audioclip.write_audiofile(
    filename = original_audio_path,
    fps = codec_audio_sr)
```

```python
# load video using MoviePy
videoclip = VideoFileClip(clip_video_path2)

input_audio_path = 'inputs/audio/input_audio.wav'

# codecがトレーニングされているsampling rateに合わせてfpsを設定
codec_audio_sr = 16000

# extract audio
audioclip = videoclip.audio
audioclip.write_audiofile(
    filename = input_audio_path,
    fps = codec_audio_sr)
```

```python
original_transcript = 'From this day forward, it\'s going to be only America First, America First.'
print(original_transcript)
display(Audio(original_audio_path, rate=codec_audio_sr))
```

```python
using_transcript = 'From this day forward, it\'s going to be'
print(using_transcript)
display(Audio(input_audio_path, rate=codec_audio_sr))
```

```python
%cd /content/VoiceCraft

!mkdir checkpoints

repo_id = 'pyp1/VoiceCraft'
hf_hub_download(
    repo_id = repo_id,
    filename = "giga830M.pth",
    local_dir = "./checkpoints")

hf_hub_download(
    repo_id = repo_id,
    filename = "encodec_4cb2048_giga.th",
    local_dir = "./checkpoints")


ckpt_fn = './checkpoints/giga830M.pth'
encodec_fn = './checkpoints/encodec_4cb2048_giga.th'
```

```python
# load checkpoint
ckpt = torch.load(ckpt_fn, map_location="cpu")
# VoiceCraft loads weights
model = voicecraft.VoiceCraft(ckpt["config"])
model.load_state_dict(ckpt["model"])
model.to(device)
model.eval()
```

```python
# load tokenizers
phn2num = ckpt['phn2num']
text_tokenizer = TextTokenizer(backend="espeak")
audio_tokenizer = AudioTokenizer(signature=encodec_fn, device=device)
```

```python
%cd /content/VoiceCraft

# using section
cut_off_sec = 4.2

# NOTE: 3 sec of reference is generally enough for high quality voice cloning,
# but longer is generally better, try e.g. 3~6 sec.
info = torchaudio.info(input_audio_path)
audio_dur = info.num_frames / info.sample_rate
print('audio_dur:', audio_dur)

assert cut_off_sec < audio_dur, f"cut_off_sec {cut_off_sec} is larger than the audio duration {audio_dur}"
prompt_end_frame = int(cut_off_sec * info.sample_rate)
print('prompt_end_frame:', prompt_end_frame)
```

```python
# the sample rate of the codec codes
codec_sr = 50
# 上位k個を選択
top_k = 0
# 上位p%を選択
top_p = 0.8
# 確率分布の確率を調整するためのパラメタ
temperature = 1
# テキストや音声データ内で無視される特定のトークン
silence_tokens=[1388,1898,131]
# 1でkv cache使用、高速化するが、消費メモリ増加。OOMの場合0に設定
kvcache = 1
# 推論時トークンの繰り返し回数上限
stop_repetition = 3
# バッチサイズ数
# 推論結果に長い無音やscratchが発生する場合、バッチ数を上げ
# 複数の選択肢から最も短いバッチを選択する
# 推論結果が極端に短い場合はバッチ数を下げる
sample_batch_size = 6
```

```python
print(original_transcript)
print(using_transcript)

target_transcript = using_transcript + ' only World peace First, Environmental conservation Second'
print(target_transcript)
```

```python
decode_config = {
    'top_k': top_k,
    'top_p': top_p,
    'temperature': temperature,
    'stop_repetition': stop_repetition,
    'kvcache': kvcache,
    "codec_audio_sr": codec_audio_sr,
    "codec_sr": codec_sr,
    "silence_tokens": silence_tokens,
    "sample_batch_size": sample_batch_size
    }

# speech editing
concated_audio, gen_audio = inference_one_sample(
    model,
    ckpt["config"],
    phn2num,
    text_tokenizer,
    audio_tokenizer,
    input_audio_path,
    target_transcript,
    device,
    decode_config,
    prompt_end_frame
    )

```

```python
# get tensor audio
concated_audio = concated_audio[0].cpu()
gen_audio = gen_audio[0].cpu()
```

```python
# Original Audio
print(original_transcript)
display(Audio(original_audio_path, rate=codec_audio_sr))
```

```python
# Text Editing Result
print(target_transcript)
display(Audio(concated_audio, rate=codec_audio_sr))
```

```python
# 生成部分のみ
print(' only World peace First, Environmental conservation Second')
display(Audio(gen_audio, rate=codec_audio_sr))
```
